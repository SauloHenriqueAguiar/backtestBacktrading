# Le API da Binance

```
import ccxt


binance = ccxt.binance ()
binance.load_markets ()

print ('binance: %d markets loaded!' % len (binance.markets))
```

# Configura o gráfico

```
# Ativa a resolução 2x
%config InlineBackend.figure_format = 'retina'

# biblioteca plotar gráfico
import matplotlib.pyplot as plt

# configura
plt.style.use ('seaborn-white')
plt.rcParams["figure.figsize"] = [15,6]
```

# Exibir preços ETH / USDT

```
import pandas as pd
from datetime import datetime

pair = 'ETH/USDT'

# Carregar dados OHLCV (abrir / alto / baixo / fechar / volume) com resolução de 1 dia
ohlcv = binance.fetch_ohlcv (pair, '1d')

# Obtenha os preços de fechamento para cada dia
prices = [x[4] for x in ohlcv]

# Converter carimbos de data / hora Unix em datas Python
dates = [datetime.fromtimestamp (x[0] // 1000) for x in ohlcv]

# Prepare um objeto da série Pandas
data = pd.Series (prices, index=dates) 

# Desenhe um gráfico
data.plot ()
```
